```python
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F

# Load model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

# Function: get sentence embedding
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state[:, 0, :]  # CLS vector

# Knowledge base (simple meanings)
knowledge = {
    "machine learning": "AI systems learn patterns from data",
    "deep learning": "Neural networks with many layers",
    "natural language processing": "AI that understands human language",
    "computer vision": "AI that understands images"
}

# Precompute embeddings
kb_embeddings = {k: get_embedding(k) for k in knowledge}

# Function: explain input
def explain_text(input_text):
    input_vec = get_embedding(input_text)

    best_score = -1
    best_match = None

    for k, v in kb_embeddings.items():
        score = F.cosine_similarity(input_vec, v).item()
        if score > best_score:
            best_score = score
            best_match = k

    return knowledge[best_match]

# Test
text = "this is the difination of bioogy "
print("Explanation:", explain_text(text))
```

```python
from moviepy.editor import VideoFileClip
import os

def get_video_info(video_path):
    # Load video
    clip = VideoFileClip(video_path)
    
    # Duration in seconds
    duration = clip.duration
    
    # File size in MB
    size_mb = os.path.getsize(video_path) / (1024 * 1024)
    
    return {
        "duration_seconds": duration,
        "size_mb": size_mb
    }

# Example usage
video_path = "sample.mp4"
info = get_video_info(video_path)

print(f"Duration: {info['duration_seconds']} seconds")
print(f"Size: {info['size_mb']:.2f} MB")
```

```python
from ultralytics import YOLO
import cv2
import numpy as np

# Load model
model = YOLO("yolov8n.pt")

video_path = "/kaggle/input/datasets/brukeahmye/common/dolarrr.mp4"
cap = cv2.VideoCapture(video_path)

# -------------------------------
# 🔹 Video metadata
# -------------------------------
fps = cap.get(cv2.CAP_PROP_FPS)
frame_count = cap.get(cv2.CAP_PROP_FRAME_COUNT)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

duration = frame_count / fps

print("📌 Video Info:")
print(f"Resolution: {width} x {height}")
print(f"FPS: {fps}")
print(f"Duration: {duration:.2f} seconds")

# Output writer
out = cv2.VideoWriter(
    "output_video.mp4",
    cv2.VideoWriter_fourcc(*"mp4v"),
    fps,
    (width, height)
)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # -------------------------------
    # 🔹 Color analysis (frame level)
    # -------------------------------
    avg_color = np.mean(frame, axis=(0, 1))  # BGR average
    b, g, r = avg_color

    color_text = f"B:{int(b)} G:{int(g)} R:{int(r)}"

    # Detect objects
    results = model(frame)

    for result in results:
        boxes = result.boxes

        for box in boxes:
            cls_id = int(box.cls[0])
            label = model.names[cls_id]

            if label == "person":
                x1, y1, x2, y2 = map(int, box.xyxy[0])

                cv2.rectangle(frame, (x1, y1), (x2, y2), (0,255,0), 2)
                cv2.putText(frame, "Human", (x1, y1 - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

    # -------------------------------
    # 🔹 Display video info on frame
    # -------------------------------
    cv2.putText(frame, f"{width}x{height}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2)

    cv2.putText(frame, f"FPS: {fps:.1f}", (10, 60),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2)

    cv2.putText(frame, f"Duration: {duration:.1f}s", (10, 90),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2)

    cv2.putText(frame, f"Avg Color: {color_text}", (10, 120),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (200,200,200), 2)

    # Save + show
    out.write(frame)
    cv2.imshow("Video Analysis", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

video_path = "/kaggle/input/datasets/brukeahmye/common/dolarrr.mp4"
cap = cv2.VideoCapture(video_path)

ret, prev_frame = cap.read()
prev_frame = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
prev_frame = cv2.GaussianBlur(prev_frame, (21, 21), 0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert current frame
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    gray = cv2.GaussianBlur(gray, (21, 21), 0)

    # Frame difference
    diff = cv2.absdiff(prev_frame, gray)

    # Threshold (detect movement)
    _, thresh = cv2.threshold(diff, 25, 255, cv2.THRESH_BINARY)

    # Count changed pixels
    motion_level = np.sum(thresh) / 255

    # ----------------------------
    # 🔔 SIMPLE NOTIFICATION LOGIC
    # ----------------------------
    if motion_level > 5000:
        cv2.putText(frame, "⚠ MOTION DETECTED!", (50, 50),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)
        print("⚠ Movement detected!")

    # Show frames
    cv2.imshow("Motion Detection", frame)
    cv2.imshow("Threshold", thresh)

    # Update previous frame
    prev_frame = gray

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
